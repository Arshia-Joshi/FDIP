```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("C:/Users/arshi/Downloads/waterlily.jpg", cv2.IMREAD_GRAYSCALE)
```

```python

if img is None:
    raise FileNotFoundError("Image not found! Please check the file path.")


min_val = np.min(img)
max_val = np.max(img)
contrast_stretched = ((img - min_val) * (255 / (max_val - min_val))).astype(np.uint8)


hist_eq = cv2.equalizeHist(img)


plt.figure(figsize=(15, 10))


plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(3, 2, 2)
plt.hist(img.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Histogram - Original")

plt.subplot(3, 2, 3)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")

plt.subplot(3, 2, 4)
plt.hist(contrast_stretched.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Histogram - Contrast Stretched")


plt.subplot(3, 2, 5)
plt.imshow(hist_eq, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

plt.subplot(3, 2, 6)
plt.hist(hist_eq.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Histogram - Equalized")

plt.tight_layout()
plt.show()
```
