```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("C:/Users/arshi/Downloads/flamingo.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
```

```python
def sample_image(image, factor):
    h, w = image.shape[:2]
    sampled = cv2.resize(image, (w//factor, h//factor), interpolation=cv2.INTER_NEAREST)
    return sampled

def quantize_image(image, levels):
    step = 256 // levels
    quantized = (image // step) * step + step // 2
    return quantized
```

```python
sampled_img = sample_image(img, 4)

quantized_img = quantize_image(img, 4)
quantized_img_2 = quantize_image(img, 8)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 5))
axs[0].imshow(img)
axs[0].set_title("Original")
axs[0].axis("off")

axs[1].imshow(sampled_img)
axs[1].set_title("Sampled (factor=4)")
axs[1].axis("off")

axs[2].imshow(quantized_img)
axs[2].set_title("Quantized (4 levels)")
axs[2].axis("off")

axs[3].imshow(quantized_img_2)
axs[3].set_title("Quantized (8 levels)")
axs[3].axis("off")
plt.savefig("output.jpg")
plt.show()
```
